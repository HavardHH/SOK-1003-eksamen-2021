```python
import pandas as pd
import numpy as np
from datetime import date
import numpy_financial as npf
import matplotlib.pyplot as plt

def verdier():
    print('Skriv inn lånebeløp: ')
    lån = float(input())
    print('Skriv inn antall år: ')
    år = float(input())
    print('Skriv inn renten som som heltall: ')
    r = float(input())
    rente = r/100
    termin_år = 12
    return lån, år, rente, termin_år
# Definerer kalkulatoren
def kalkulator (rente, år, termin_år, lån, start_date=date.today()):
    rng = pd.date_range(start_date, periods=år*termin_år, freq='MS') # MS står for måned start
    rng.name = 'Betalingsdato'

    # Lager dataframen som skal printe ut lånet etter terminer
    df = pd.DataFrame(index=rng, columns=['Terminbeløp', 'Lån', 'Rente', 'Restlån'], dtype='float')

    # Lager en index for terminene som starter på 1 og ikke 0
    df.reset_index(inplace=True)
    df.index += 1
    df.index.name = 'Termin'

    #Lager utregningene til selve kalkulatoren, numpy_financial har disse funksjonene innebygd
    terminebløp = npf.pmt(rente/termin_år, år*termin_år, lån)
    df['Terminbeløp'] = -terminebløp
    df['Lån'] = -npf.ppmt(rente/termin_år, df.index, år*termin_år, lån)
    df['Rente'] = -npf.ipmt(rente/termin_år, df.index, år*termin_år, lån)

    # For å avrunde tallene til bare 2 desimaler bruker vi funksjonen round.
    df = df.round(2)

    # Definerer hvor mye lån vi har betalt og passe på den ikke går over det orginale lånebeløpet
    df['Betalt lån'] = df['Lån'].cumsum()
    df['Betalt lån']  = df['Betalt lån'].clip(lower=-lån)

    # Definerer restlånet og summen av betalt rente
    df['Restlån'] = lån - df['Betalt lån']
    df['Sum renter'] = df['Rente'].abs().cumsum()

    # Definerer den siste betalingsdatoen
    try:
        siste_betaling = df.query('Restlån <= 0')['Restlån'].idxmax(axis=1, skipna=True)
    except ValueError:
        siste_betaling = df.last_valid_index()

    siste_betaling_dato = "{:%m-%d-%Y}".format(df.loc[siste_betaling, "Betalingsdato"])

    # Får betalingsinfoen inn i ett DataFrame
    betaling_info = (df[["Terminbeløp", "Lån", "Rente"]].sum().to_frame().T)

    betaling_detaljer = pd.DataFrame.from_dict(dict([('nedbetaling', [siste_betaling_dato]),
                                                     ('Renten', [rente]),
                                                     ('Antall år', [år])
                                                     ]))

    # Lager en kolonne med hvor mye vi betaler hver måned
    betaling_detaljer["Terminer"] = round(terminebløp, 2)

    Betaling_sammendrag = pd.concat([betaling_detaljer, betaling_info], axis=1)
    return df, Betaling_sammendrag

Tabell, stats1 = kalkulator(0.05, 20, 4, 1000000)
pd.set_option('display.max_rows', Tabell.shape[0]+1)

display(Tabell)

fig, ax = plt.subplot()
ax.bar( x="Betalingsdato", y="rente", ax=ax)
plt.title("Rente betalinger")
```
